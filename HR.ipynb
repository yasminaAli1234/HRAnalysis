```python
# convert BirthDate,HireDate,RateChangeDate to date type
Hr['BirthDate'] = pd.to_datetime(Hr['BirthDate'])
Hr['HireDate'] = pd.to_datetime(Hr['HireDate'])
Hr['RateChangeDate'] = pd.to_datetime(Hr['RateChangeDate'])
```

```python
# convert EndTime ,StartTime to date type
Hr['EndTime'] = pd.to_datetime(Hr['EndTime'])
Hr['StartTime'] = pd.to_datetime(Hr['StartTime'])
```

```python
# round age
Hr['Age'] = round(Hr['Age'])
```

```python
Hr['ShiftDuration'] = (Hr['EndTime'] - Hr['StartTime']).dt.total_seconds() / 3600
```

```python
Hr['SalaryLevel'] = pd.qcut(Hr['Salary'], q=3, labels=['Low','Medium','High'])
```

```python
def categorize_job(title):
    title = title.lower()
    if 'chief' in title or 'vice president' in title:
        return 'Executive / Leadership'
    elif 'manager' in title or 'supervisor' in title:
        return 'Management / Supervisory'
    elif 'engineer' in title or 'r&d' in title or 'designer' in title or 'control specialist' in title:
        return 'Engineering / Technical'
    elif 'production' in title or 'technician' in title or 'quality' in title or 'stocker' in title or 'shipping' in title:
        return 'Production / Manufacturing'
    elif 'sales' in title or 'marketing' in title:
        return 'Sales / Marketing'
    elif 'account' in title or 'finance' in title or 'purchasing' in title or 'buyer' in title:
        return 'Finance / Accounting'
    elif 'human resources' in title or 'recruiter' in title or 'benefits' in title:
        return 'Human Resources'
    elif 'network' in title or 'database' in title or 'application' in title or 'information services' in title:
        return 'IT / Information Systems'
    elif 'assistant' in title or 'scheduler' in title:
        return 'Administration / Support'
    elif 'facilities' in title or 'maintenance' in title or 'janitor' in title:
        return 'Facilities / Maintenance'
    else:
        return 'Other'

Hr['JobCategory'] = Hr['JobTitle'].apply(categorize_job)
```
